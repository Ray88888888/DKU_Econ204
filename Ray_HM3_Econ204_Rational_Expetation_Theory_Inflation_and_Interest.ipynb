# Task 1: The Monetary Market

$M^d/P=m*Y/i$

In equilibrum $M^s=M^d$

Please Change one the exogenous variable for your homework

```
### Define exogenous variables
Ms= 1000
m=1
Y=20
i=0.05
```

```
### Calculate the equilibrium P level
P=Ms*i/(m*Y)
print(P)
```

```
### draw the equilibrium
import matplotlib.pyplot as plt
import numpy as np
fig= plt.figure(figsize=(20,10))
x = np.linspace(0, 2000)
y = i/(m*Y)*x
###draw an vertical line for monetary supply
plt.axvline(x=Ms, Label='Ms')
plt.plot(x,y, Label='Md=L(Y,i)')

###plot label, legend, and title
plt.xlabel('M')
plt.ylabel('P')
plt.legend()
plt.title('The Monetary Market Equlibrium')
plt.show()
```
